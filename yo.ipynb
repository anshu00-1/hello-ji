```python
class LibraryItem:
    def __init__(self,title, author,publication_year):
        self.title=title
        self.author=author
        self.publication_year=publication_year

    def book(self):
        print(f"THIS BOOK IS WRITTEN BY {self.author} AND IT WAS PUBLISHED IN YEAR {self.publication_year}")

    def magazine(self):
        print(f"THIS MAGAZINE IS WRITTEN BY {self.author} AND IT WAS PUBLISHED IN YEAR {self.publication_year}")

class taken_time:
    def __init__(self,checkout,return_item):
        self.checkout=checkout
        self.return_item=return_item

    def time_book(self):
        print(f"DATE FOR ISSUENCE OF THIS BOOK IS {self.checkout} AND THE DATE FOR RETURN WILL BE {self.return_item}")

    def time_magazine(self):
        print(f"DATE FOR ISSUENCE OF THIS MAGAZINE IS {self.checkout} AND THE DATE FOR RETURN WILL BE {self.return_item}")


        
book1=LibraryItem("48 LAWS OF POWER","ROBERT GREENE","1987")
book2=taken_time("26/04/2006","16/07/2006")

magazine1=LibraryItem("kya pata","ROBERT GREENE","1987")
magazine2=taken_time("26/04/2006","16/07/2006")

book1.book()
book2.time_book()

magazine1.magazine()
magazine2.time_magazine()
```

```python
class Person:
    def __init__(self,name ,age ,address):
        self.name=name
        self.age=age
        self.address=address

class Student(Person):
    def __init__(self,name ,age ,address):
        super().__init__(name ,age ,address)

    def take_exam(self,take_exam):
        self.take_exam=take_exam
        print(f"{self.name} is taking the exam? {self.take_exam}")
        print(f"what is his age: {self.age}")

class Teacher(Person):
    def __init__(self,name ,age ,address):
        super().__init__(name ,age ,address)
        
    def teach(self,teach_subject):
        self.teach_subject=teach_subject
        print(f"{self.name} is teaching? {self.teach_subject}")

class Staff(Person):
    def __init__(self,name ,age ,address):
        super().__init__(name ,age ,address)
        print(f"{self.name} is taking the exam: {self.age}")

person1=Student("ayush",17,"yahi kahi")
person1.take_exam("na mai nahi de raha")

person1=Teacher("mahipal",30,"yahi kahi")
person1.teach("na mai nahi padha raha")

person1=Staff("ayush",17,"yahi kahi")
```

```python
# odd even numbers

number=(int(input("YO ENTER A NUMBER HERE")))

if number%2==0:
    print(f"{number} IS AN EVEN NUMBER")

else:
    print(f"{number} IS AN ODD NUMBER")
```

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
number = int(input("Enter a number: "))

if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```
